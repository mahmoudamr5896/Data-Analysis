```python
avocado_ts = avocado[['Date','AveragePrice']]
```

```python
avocado_ts.columns = ['ds','y']
```

```python
avocado_ts
```

```python
from prophet import Prophet
```

```python

m = Prophet()
```

```python
m.fit(avocado_ts)
```

```python
future= m.make_future_dataframe(periods=365)
```

```python
forcast = m.predict(future)
```

```python
forcast
```

```python
fig_1 =m.plot(forcast)
fig_1
```

```python
fig_2= m.plot_components(forcast)
fig_2
```
